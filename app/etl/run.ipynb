# Celo Alfajores 交易數據分析

這個 notebook 用於分析 Celo Alfajores 測試網的交易數據。

```python
import requests
import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns

# 設置顯示選項
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)
# plt.style.use('seaborn')
```

```python
from datetime import datetime

BASE_URL = 'https://celo-alfajores.blockscout.com/api/v2/transactions?filter=validated&type=token_transfer'
block_number = None
index = None
items_count = None

# Get 1000 items
all_items = []

while len(all_items) < 1000:
    if block_number is not None:
        url = f"{BASE_URL}&block_number={block_number}&index={index}&items_count={items_count}"
    else:
        url = BASE_URL

    print(url)

    res = requests.get(url).json()
    block_number = res['next_page_params']['block_number']
    index = res['next_page_params']['index']
    items_count = res['next_page_params']['items_count']

    all_items.extend(res['items'])

df = pd.DataFrame(all_items)

# 生成包含時間戳的文件名
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f"transactions_{timestamp}.csv"

# 保存到 CSV 文件
df.to_csv(filename, index=False)
print(f"\nData saved to {filename}")

# # 獲取數據
# res = requests.get(URL).json()
# df = pd.DataFrame(res['items'])

# # 顯示數據基本信息
# print("數據形狀:", df.shape)
# # print("\n數據列:")
# # print(df.columns.tolist())
# # print("\n數據預覽:")
# df.head()
```

```python
import ast
df = pd.read_csv('transactions_20250530_225448.csv')

# df['from']
df['from_hash'] = df['from'].apply(lambda x: ast.literal_eval(x)['hash'] if isinstance(x, str) else None)
df['to_hash'] = df['to'].apply(lambda x: ast.literal_eval(x)['hash'] if isinstance(x, str) else None)

df['from_hash'] = df['from_hash'].map(df['from_hash'].value_counts())
df['to_hash'] = df['to_hash'].map(df['to_hash'].value_counts())
# df['to_hash'] = df['to'].apply(lambda x: x['hash'] if isinstance(x, dict) and 'hash' in x else None)
# print(df.columns.tolist())

df
```

```python
from sklearn.ensemble import IsolationForest

features = ['from_hash', 'to_hash', 'value']
X = df[features]

model = IsolationForest(n_estimators=50, contamination=0.01, random_state=42)
model.fit(X)
```

```python
df['anomaly'] = model.predict(X)
df[df['anomaly'] == -1][features]
```
